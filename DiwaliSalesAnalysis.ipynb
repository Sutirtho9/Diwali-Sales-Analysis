```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

# <span style="color:purple;">Cleaning Data</span>

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

# <span style="color:green;">Exploratory Analysis</span>

## <span style="color:red;">Gender</span>

```python
ax=sns.countplot(x='Gender', hue='Gender', data=df, palette=['pink', 'blue'], legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender' , hue ='Gender',y='Amount',data= sales_gen, palette=['pink','blue'], legend=False)
```

## <span style="color:red">Age</span>

```python
ax=sns.countplot(x='Age Group' ,hue = 'Gender',data=df, palette=['pink','blue'], legend=True)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',hue='Amount',y='Amount', data=sales_age, palette='Blues_d')
```

## <span style='color:red'>State</span>

```python
ord_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x='State',hue='State',y='Orders', data=ord_state, palette='RdBu_d')
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x='State',hue='State',y='Amount', data=sales_state, palette='pastel')
```

## <span style='color:red'>Marital Status</span>

```python
ax=sns.countplot(x='Marital_Status', hue ='Marital_Status', data=df , palette=['red','green'] ,legend=False)
sns.set(rc={'figure.figsize':(1,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_mart=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(8,7)})
sns.barplot(x='Marital_Status',hue='Gender',y='Amount',data=sales_mart,palette=['pink','blue'])
```

## <span style='color:red'>Ocupation</span>

```python
ax=sns.countplot(x='Occupation' ,data=df ,hue='Occupation', palette='pastel')
sns.set(rc={'figure.figsize':(25,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,8)})
sns.barplot(x='Occupation',hue='Occupation',y='Amount',data=sales_occ,palette='pastel')
```

## <span style='color:red'>Product Category</span>

```python
ax=sns.countplot(x='Product_Category' ,data=df ,hue='Product_Category', palette='pastel')
sns.set(rc={'figure.figsize':(25,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(30,8)})
sns.barplot(x='Product_Category',hue='Product_Category',y='Amount',data=sales_occ,palette='pastel')
```
